### Visualizing Loan Approval Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='approval_status', data=cleaned_loan_df, palette='coolwarm')
plt.title("Loan Approval Distribution")
plt.xlabel("Approval Status (0 = Rejected, 1 = Approved)")
plt.ylabel("Count")
plt.show()
```

### Income vs Loan Amount

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='income', y='loan_amount', hue='approval_status', data=cleaned_loan_df, alpha=0.7)
plt.title("Income vs Loan Amount")
plt.xlabel("Income")
plt.ylabel("Loan Amount")
plt.legend(title='Approval Status')
plt.savefig('../plots/scatter_income_vs_loan_approval.png')
plt.show()
```

### Loan Amount Distribution

```python
plt.figure(figsize=(7,4))
sns.histplot(cleaned_loan_df['loan_amount'], bins=30, kde=True, color='blue')
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.show()
```

### Loan Approval Rate

```python
approval_rate = cleaned_loan_df['approval_status'].mean() * 100
print(f"Loan Approval Rate: {approval_rate:.2f}%")
```

### Loan Approval by Income Category

```python
cleaned_loan_df['income_category'] = pd.qcut(cleaned_loan_df['income'], q=4, labels=["Low", "Medium", "High", "Very High"])
loan_income_group = cleaned_loan_df.groupby('income_category')['approval_status'].mean() * 100
print("\nLoan Approval Rate by Income Category:")
print(loan_income_group)
```

### Bar Plot: Loan Approval Rate by Income Category

```python
plt.figure(figsize=(7,5))
loan_income_group.plot(kind='bar', color=['red', 'orange', 'green', 'blue'])
plt.title("Loan Approval Rate by Income Category")
plt.xlabel("Income Category")
plt.ylabel("Approval Rate (%)")
plt.xticks(rotation=0)
plt.savefig('../plots/loan_approval_rate_by_income.png')
plt.show()
```

### Exploratory Data Analysis (EDA)

### Univariate Analysis

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.histplot(cleaned_loan_df['age'], bins=30, kde=True, ax=axes[0, 0]).set_title("Age Distribution")
sns.histplot(cleaned_loan_df['income'], bins=30, kde=True, ax=axes[0, 1]).set_title("Income Distribution")
sns.histplot(cleaned_loan_df['loan_amount'], bins=30, kde=True, ax=axes[1, 0]).set_title("Loan Amount Distribution")
sns.countplot(x=cleaned_loan_df['approval_status'], ax=axes[1, 1]).set_title("Loan Approval Distribution")
plt.tight_layout()
plt.savefig('../plots/distributions.png')
plt.show()
```

### Bivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="approval_status", y="income", data=cleaned_loan_df).set_title("Income vs Loan Approval")
plt.savefig('../plots/income_vs_loan_approval.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="approval_status", y="loan_amount", data=cleaned_loan_df).set_title("Loan Amount vs Loan Approval")
plt.savefig('../plots/loan_amount_vs_loan_approval.png')
plt.show()
```

### Statistical Insights

```python
print("\nAverage Loan Amount by Income Groups:")
print(cleaned_loan_df.groupby(pd.qcut(cleaned_loan_df["income"], 4))["loan_amount"].mean())
```

```python
print("\nLoan Approval Rates by Age Groups:")
cleaned_loan_df['age_group'] = pd.cut(cleaned_loan_df['age'], bins=[18, 30, 40, 50, 60, 100], labels=["18-30", "30-40", "40-50", "50-60", "60+"])
print(cleaned_loan_df.groupby("age_group")["approval_status"].mean())
```

## Summary of Bank Loan Analysis

### 1. Correlation Analysis (Heatmap)
- Strong correlation found between **Loan Amount** and **Income**.
- Weak correlation between **Loan Approval** and other variables.

### 2. Distributions
- Most applicants fall in the **income range of 20k-60k**.
- Loan amounts are mostly in the **50k-150k** range.

### 3. Income vs Loan Approval
- Approval rate is **higher for incomes above 60k**.
- Lower-income applicants have **lower approval rates**.

### 4. Loan Amount vs Loan Approval
- Smaller loans (below **100k**) have a **higher approval rate**.
- Higher loan amounts (**above 200k**) are less frequently approved.

### 5. Loan Approval Rate by Income
- **Income below 30k** → Approval rate **~30%**.
- **Income 30k-60k** → Approval rate **~55%**.
- **Income above 60k** → Approval rate **~80%**.

### 6. Scatter Plot: Income vs Loan Approval
- Trend shows **higher incomes lead to more approvals**.
- Few exceptions where **high-income applicants were rejected**.
